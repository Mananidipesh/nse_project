```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('D://nse//nse_data2.csv')
df
```

```python
df.info()
```

```python
for column in df:
    if column == 'name':
        pass
    elif df[column].dtype == object:
        df[column] = df[column].str.replace(',', '')
        df[column] = df[column].astype(float)
```

```python
df.info()
```

```python
df.describe().T
```

```python
#Line graph chart for Nifty fifty
# Prepare x-axis labels
x_labels = ['open', 'high', 'low', 'L.T.P']

# Prepare y-axis values
y_values = [df['open'][0], df['high'][0], df['low'][0], df['L.T.P'][0]]

# Plotting
plt.figure(figsize=(10, 5))

# Plot Open, High, Low, and LTP
plt.plot(x_labels, y_values, marker='o', label='Prices', color='blue')

# Adding labels and title
plt.title('Daily Price Movement for ' + df['name'][0], fontsize=16)
plt.xlabel('Price Type', fontsize=14)
plt.ylabel('Price ()', fontsize=14)
plt.xticks(rotation=0)
plt.grid()

# Adding values on the points for clarity
for i, value in enumerate(y_values):
    plt.text(x_labels[i], value + 1, str(value), ha='center', fontsize=12)

# Show the plot
plt.show()
```

```python
#Comparison bar chart for low and high prices for stocks 
# Set the bar width and positions
bar_width = 0.35
x = np.arange(len(df['name']))

# Create the plot
plt.figure(figsize=(10, 10))
# Create bars for each dataset
plt.barh(df['name'], df['low'], label='Dataset 1', color='blue')
plt.barh(df['name'], df['high'], left=df['low'], label='Dataset 2', color='orange')

# Adding titles and labels
plt.title('Comparison of low & high price of stocks')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.xticks(x, df['name'])  # Set x-ticks to the category names
plt.legend()
plt.grid(axis='y')


# Show the plot
plt.tight_layout()
plt.show()
```
